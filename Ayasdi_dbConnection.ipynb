```python
""" Getting list of tables"""

# Importing libraries
import os
import codecs

# Source file
file_path = 'tables.txt'

# Opening list of tables
tables = codecs.open(file_path, 'r')

    
""" Connecting to database """

# Importing libraries
import sqlite3

# Connecting to db
db_path = 'database.db'
conn = sqlite3.connect(db_path)
# Instantiating cursor
c = conn.cursor()

# Getting folder from datase file name
folder = os.path.dirname(db_path)

# Getting database file name
db_name = os.path.basename(db_path)
db_name = os.path.splitext(db_name)[0]


""" Extracting tables to csv files and loading to Ayasdi """

# Importing libraries
from ayasdi_sdk import *

# Browsing list of tables
for tb_name in tables:
    # Querying database
    sql = 'SELECT * FROM ' + tb_name
    
    # Fetching query
    d = c.execute(sql)
    
    # Exporting records to csv file
    if d is not None:
        # Generating target file name
        out_file_path = ''
        if folder != '': out_file_path = folder + os.path.sep
        out_file_path = out_file_path + db_name + "_" + tb_name.rstrip() + '.csv'
        
        # Closing target file if needed
        try: os.unlink(out_file_path)
        except OSError: pass
        
        # Upating target file from table
        out_file = codecs.open(out_file_path, 'w')
        r = 0
        for row in d:
            line = ''
            r += 1
            for k in range(len(row)):
                line = line + str(row[k]) + ','
            
            # Append record to target file
            out_file.write(line[:-1] + "\n")
            
        # Closing target file
        out_file.close()
        print r, ' rows populated in file ', out_file_path
        
        # Loading to Ayasdi
        ayasdi_loader(out_file_path)
        
# Closing list of tables
tables.close()
```
